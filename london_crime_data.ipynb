```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import requests
```

```python
data_folder = 'data'
output_folder = 'output'

if not os.path.exists(data_folder):
    os.mkdir(data_folder)
if not os.path.exists(output_folder):
    os.mkdir(output_folder)
```

```python
def download(url):
    filename = os.path.join(data_folder, os.path.basename(url))
    if not os.path.exists(filename):
      with requests.get(url, stream=True, allow_redirects=True) as r:
          with open(filename, 'wb') as f:
              for chunk in r.iter_content(chunk_size=8192):
                  f.write(chunk)
      print('Downloaded', filename)
```

```python
files = [
  '2020-01-metropolitan-street.csv',
  '2020-02-metropolitan-street.csv',
  '2020-03-metropolitan-street.csv',
  '2020-04-metropolitan-street.csv',
  '2020-05-metropolitan-street.csv',
  '2020-06-metropolitan-street.csv',
  '2020-07-metropolitan-street.csv',
  '2020-08-metropolitan-street.csv',
  '2020-09-metropolitan-street.csv',
  '2020-10-metropolitan-street.csv',
  '2020-11-metropolitan-street.csv',
  '2020-12-metropolitan-street.csv'
]


data_url = 'https://github.com/spatialthoughts/python-dataviz-web/releases/' \
  'download/police.uk/'

for f in files:
  url = os.path.join(data_url + f)
  download(url)
```

```python
dataframe_list = []

for f in files:
    filepath = os.path.join(data_folder, f)
    df = pd.read_csv(filepath)
    dataframe_list.append(df)

merged_df = pd.concat(dataframe_list)
```

```python
merged_df
```

```python
type_counts = merged_df.groupby('Crime type').size()
type_counts
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,10)

type_counts.plot(kind='pie', ax=ax)

ax.set_title('Crime Types', fontsize = 18)
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
wedgeprops={'linewidth': 3, 'edgecolor': 'white'}
textprops= {'fontsize': 10, 'fontstyle': 'italic'}

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,10)

type_counts.plot(kind='pie', ax=ax,
                 wedgeprops=wedgeprops,
                 textprops=textprops
                 )

ax.set_title('Crime Types', fontsize = 18)
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
monthly_counts = merged_df.groupby('Month').size()
monthly_counts
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,7)
monthly_counts.plot(kind='bar', ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,7)

monthly_counts.plot(kind='bar', ax=ax)
monthly_counts.plot(kind='line', ax=ax, color='red', marker='o')

ax.set_title('Total Crime by Month')
ax.set_ylabel('Total Incidents')

plt.show()
```

```python
fig, axes = plt.subplots(7, 2, figsize=(30, 30))  # Create a 7x2 grid of subplots
axes = axes.flatten()  # Flatten the 2D array of axes for easier iteration

for i, crime_type in enumerate(merged_df['Crime type'].unique()):
    if i < len(axes):  # Check to prevent exceeding subplot count
        ax = axes[i]  # Select the appropriate subplot
        crime_df = merged_df[merged_df['Crime type'] == crime_type]
        monthly_counts = crime_df.groupby('Month').size()
        monthly_counts.plot(kind='line', ax=ax, color='red', marker='o')
        ax.set_title(crime_type)
        ax.set_xlabel("Month")
        ax.set_ylabel("Number of Crimes")
    else:
        print("Too many crime types for the subplot grid!")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
import seaborn as sns

sns.set_theme()  # Apply Seaborn's theme for aesthetics

# Assuming `monthly_counts` is already a pandas Series with months as the index
fig, ax = plt.subplots(figsize=(15, 7))  # Create a single subplot with specified size

# Plot the bar chart
bar_plot = monthly_counts.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')

# Set titles and labels
ax.set_title('Total Crime by Month', fontsize=16)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Total Incidents', fontsize=12)

# Add annotations with padding and boxes at the top of each bar
for p in bar_plot.patches:
    height = p.get_height()
    ax.annotate(
        f'{int(height)}',  # Text to display
        (p.get_x() + p.get_width() / 2, height + 2),  # Position with added padding
        ha='center', va='bottom',  # Center alignment
        fontsize=10,  # Font size
        bbox=dict(
            boxstyle="round,pad=0.3",  # Rounded box style with padding
            edgecolor="black",  # Black border for the box
            facecolor="lightyellow"  # Background color of the box
        )
    )

# Improve x-axis readability
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
